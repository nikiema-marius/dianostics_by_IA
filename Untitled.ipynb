```python
import pandas as pd
```

```python
data = pd.read_csv("insurance.csv")
print("le nombre de ligne est {}",format(data.shape[0]))
data.head()
```

```python
data.describe()
data.duplicated().sum()
```

```python
data.drop_duplicates(inplace=True)
data.duplicated().sum()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

data.value_counts("sex")
```

```python
sns.countplot(x="sex", data = data)
plt.show()
```

```python
sns.countplot(x = "smoker", data = data)
plt.show()
```

```python
sns.countplot(x = "region", data = data)
plt.show()
```

```python
from sklearn import preprocessing
```

```python
labelencoder = preprocessing.LabelEncoder()
data.head()
```

```python
#fit_transform permet de transformation des donns chaine de caractere en numbre 
data["smoker"] = labelencoder.fit_transform(data["smoker"])
data["sex"] = labelencoder.fit_transform(data["sex"])
data["region"] = labelencoder.fit_transform(data["region"])
```

```python
data.head()
```

```python
data.corr()["charges"].sort_values(ascending=False)
```

```python
#permet de voire sous forme matricielle les correlation (relation) entre les donnes
sns.heatmap(data.corr(),annot = True, cmap = "rainbow")
```

```python
#analyse des distribution qu'il exsiste entre les colonnes 
for i in data.columns:
    sns.displot(data[i])
    plt.title(i + " "+ "distribution")
    plt.show()
```

```python
colorsforx = ["red", "Green", "yellow", "Black", "Blue", "Grey"]
colorchanger = 0
```

```python
# nuage des points, relation entre xet ypeut tre affiche pour diffrents sous-ensembles de donnes 
for i in data.columns[:-1]:
    sns.scatterplot(data = data, x= i, y= "charges", color =  colorsforx[colorchanger])
    plt.show()
    colorchanger += 1
```
